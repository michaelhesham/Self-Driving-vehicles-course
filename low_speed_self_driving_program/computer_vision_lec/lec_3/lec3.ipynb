```python
import cv2
import numpy as np

def nothing():
    pass

img = np.zeros((500,500),np.uint8)

cv2.namedWindow('Image')

cv2.createTrackbar('row','Image',0,img.shape[0]-1,nothing)
cv2.createTrackbar('column','Image',0,img.shape[1]-1,nothing)
cv2.createTrackbar('x','Image',0,img.shape[0]-1,nothing)
cv2.createTrackbar('y','Image',0,img.shape[1]-1,nothing)

while 1:
    row =cv2.getTrackbarPos('row','Image')
    column =cv2.getTrackbarPos('column','Image')
    x =cv2.getTrackbarPos('x','Image')
    y =cv2.getTrackbarPos('y','Image')
    
    img = np.zeros((500,500),np.uint8)
    img[row,:] = 255
    img[:,column] = 255
    img[x,y] = 255

    cv2.imshow('Image',img)

    k = cv2.waitKey(1)
    if k == 27:
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img1 = np.zeros((200,200),np.uint8)
plt.imshow(img1, cmap='gray')
```

```python
np.fill_diagonal(img1,255)
plt.imshow(img1, cmap='gray')
```

```python
img1 = np.flipud(img1)
img1[10,:] = 255
plt.imshow(img1, cmap='gray')
```

```python
img1 = np.fliplr(img1)
plt.imshow(img1, cmap='gray')
```
