```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('circles.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blur = cv2.medianBlur(img_gray,5)

def nothing(x):
    pass

cv2.namedWindow('output')
cv2.createTrackbar('x', 'output', 1, 100, nothing)
cv2.createTrackbar('y', 'output', 19, 100, nothing)
cv2.createTrackbar('param1', 'output', 100, 100, nothing)
cv2.createTrackbar('param2', 'output', 19, 100, nothing)
cv2.createTrackbar('minR', 'output', 18, 100, nothing)
cv2.createTrackbar('maxR', 'output', 60, 100, nothing)

while 1:
    cimg = img

    x = cv2.getTrackbarPos('x', 'output')
    y = cv2.getTrackbarPos('y', 'output')
    param1 = cv2.getTrackbarPos('param1', 'output')
    param2 = cv2.getTrackbarPos('param2', 'output')
    minR = cv2.getTrackbarPos('minR', 'output')
    maxR = cv2.getTrackbarPos('maxR', 'output')

    if x < 1:
        x = 1
    if y < 1:
        y = 1
    if param1 < 1:
        param1 = 1
    if param2 < 1:
        param2 = 1
    if minR > maxR:
        minR = maxR - 1

    circles = cv2.HoughCircles(img_blur, cv2.HOUGH_GRADIENT, x, y,
                            param1=param1,param2=param2,minRadius=minR,maxRadius=maxR)

    if circles is not None:
        circles = np.uint16(np.around(circles))

        for i in circles[0,:]:
            # draw the outer circle
            cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
            # draw the center of the circle
            cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)

    cv2.imshow('output', cimg)
    
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
plt.figure(figsize=(10,24))
plt.imshow(cimg[:,:,::-1])
# parameters
# x=1, y=19, param1=100, param2=19, minR=18, maxR=60
```
