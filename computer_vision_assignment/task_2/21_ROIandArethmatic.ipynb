## Import cv2, numpy, matplotlib

```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt 
```

## read image

```python
img = cv2.imread('tekomoro.png')
plt.imshow(img[:,:,::-1]);
img.shape
```

## resize logo width to 150px with reserved ratio

```python
ratio = 150/1780

img = cv2.resize(img,None,fx=ratio, fy=ratio, interpolation = cv2.INTER_LINEAR)
plt.imshow(img[:,:,::-1]);
img.shape
```

## translate logo 10px away from right and bottom sides.

```python
M = np.float32([
	[1, 0, -10],
	[0, 1, -10]
])
img2 = cv2.warpAffine(img, M, (img.shape[1],img.shape[0]))
plt.imshow(img2[:,:,::-1]);
img.shape
```
