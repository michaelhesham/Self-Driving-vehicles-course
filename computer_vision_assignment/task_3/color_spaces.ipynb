# Task 1

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
img = cv2.imread('13_output.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)

```

## Add 10 to all S channel values.

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
s = img_hsv[:,:,1] + 10
img_hsv[:,:,1] = s
img_s = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2RGB)
plt.imshow(img_s)
```

image colors are more saturated

you cannot display HSV image using imshow as it expects RGB image

## Subtract 5 from H channel.

```python
h = img_hsv[:,:,0] - 5
img_hsv[:,:,0] = h
img_h = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2RGB)
plt.imshow(img_h)
```

Slightly shift in image colors

## Myltiply V channel by 1.25.

```python
v = img_hsv[:,:,2]
for i in range(768):
    for j in range(1365):
        if v[i][j] < 200:
            v[i][j] *= 1.25
        else:
            v[i][j] = 255

img_hsv[:,:,2] = v
img_v = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2RGB)
plt.imshow(img_v)
```

Brightened parts of image become black as it become more than 255 and become black

to solve this problem we made 2 for loop to check if value < 200 multiply  v channel by 1.25 else v channel equal 255

# Task 2

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

## Change image from BGR to yCrCb

```python
img2 = cv2.imread('13_output.png')
img2_yCrCb = cv2.cvtColor(img2, cv2.COLOR_BGR2YCrCb)
```

## Multiply Y channel by 1.25 in YCrCb colorspace.

```python
y = img2_yCrCb[:,:,0]
for i in range(768):
    for j in range(1365):
        if y[i][j] < 200:
            y[i][j] *= 1.25
        else:
            y[i][j] = 255
img2_yCrCb[:,:,0] = y
img2_y = cv2.cvtColor(img2_yCrCb,cv2.COLOR_YCrCb2RGB)
plt.imshow(img2_y)
```

Brightened parts in image become black as what happend when v channel multiplyed by 1.25 and solved as in v channel
