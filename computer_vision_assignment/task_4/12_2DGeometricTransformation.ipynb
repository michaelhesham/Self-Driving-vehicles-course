```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
```

## Show image to get x, y coordinates of wanted points

```python
img1 = cv2.imread('perspective02.jpg')

img1 = cv2.resize(img1, None, fx=0.5, fy=0.5)

while 1:
    cv2.imshow('image',img1)

    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

## prespective transformation

```python
while 1:

    pts1 = np.float32(
        [
            [292,  172], [353,  172],
            [0, 359], [639, 359]
        ]
    )

    pts2 = np.float32(
        [
            [292,   0], [353,  0],
            [292, 359], [353, 359]
        ]
    )

    rows,cols = img1.shape[:2]

    M = cv2.getPerspectiveTransform(pts1 ,pts2)
    dst = cv2.warpPerspective(img1, M, (cols,rows))

    dst = cv2.resize(dst, None, fx=2, fy=2)
    cv2.imshow('Image', dst)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
plt.imshow(dst[:,:,::-1])
cv2.destroyAllWindows()
```

## Show image to get x, y coordinates of wanted points

```python
img2 = cv2.imread('perspective01.jpg')

img2 = cv2.resize(img2, None, fx=0.5, fy=0.5)

while 1:
    cv2.imshow('image2',img2)

    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

## prespective transformation

```python
while 1:

    pts3 = np.float32(
        [
            [175,  85], [280,  85],
            [0, 127], [394, 127]
        ]
    )

    pts4 = np.float32(
        [
            [175,   0], [280,  0],
            [175, 269], [280, 269]
        ]
    )

    rows2, cols2 = img2.shape[:2]

    M2 = cv2.getPerspectiveTransform(pts3 ,pts4)
    dst2 = cv2.warpPerspective(img2, M2, (cols2,rows2))

    dst2 = cv2.resize(dst2, None, fx=2, fy=2)
    cv2.imshow('Image2', dst2)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
plt.imshow(dst2[:,:,::-1])
cv2.destroyAllWindows()
```
