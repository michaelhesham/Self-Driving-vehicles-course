```python
import cv2
from matplotlib import pyplot as plt
import numpy as np 
```

```python
img = cv2.imread('lab.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_inv = cv2.bitwise_not(gray)

_, th = cv2.threshold(gray_inv, 10, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

while 1:
    out = np.ones((img.shape[0], img.shape[1], 3), np.uint8)
    out.fill(255)
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 30000:
            cv2.fillPoly(out, pts =[contour], color=(0,0,255))
        elif area > 1000:
             cv2.fillPoly(out, pts =[contour], color=(255,0,0))
    
    cv2.imshow('Output', out)

    k = cv2.waitKey(0)
    if k == 27:
        break
cv2.destroyAllWindows()
plt.figure(figsize=(24,10))
plt.imshow(out[:,:,::-1]);
```
